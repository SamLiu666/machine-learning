# 线性回归从零实现

```python
%matplotlib inline
import tensorflow as tf
print(tf.__version__)
from matplotlib import pyplot as plt
import random
```

```python
num_input = 2
num_examples = 1000
tw = [2, -1]
tb = 2.2

features = tf.random.normal((num_examples, num_input), stddev=1,dtype=tf.float32)
labels = tw[0] * features[:, 0] + tw[1]*features[:,1] + tb
```

```python
labels += tf.random.normal(labels.shape,stddev=0.01)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(features[:,0], labels, 1)  # first param
plt.scatter(features[:,1], labels, 2)  # secend param
```

```python
def data_iter(batch_size, features, labels):
    # 批量读取数据
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    
    for i in range(0, num_examples, batch_size):
        j = indices[i: min(i+batch_size, num_examples)]
        yield tf.gather(features, axis=0, indices=j), tf.gather(labels, axis=0, indices=j)
```

```python
batch = 8
for X,y in data_iter(batch, features, labels):
    print(X,y)
    break
```

```python
# 初始化模型参数
w = tf.Variable(tf.random.normal((num_input, 1), stddev=0.01))
b = tf.Variable(tf.zeros((1,)))

def linearg(X, w, b):
    # 模型定义
    return tf.matmul(X, w)+b

def squared_loss(y_hat, y):
    # 损失函数定义
#     s = y_hat.shape
    return  (y_hat - tf.reshape(y, y_hat.shape))**2 /2

def sgd(params, lr, batch, grads):
    # 批量随机梯度下降
    for i, param in enumerate(params):
        param.assign_sub(lr * grads[i] / batch)

# a = linearg(X,w,b)
# print(a)
# b = labels.shape
# print(b)
# s = tf.reshape(a, b)
# # print(s)
```

```python
# 训练模型
lr = 0.03
num_epochs = 6
net = linearg
loss = squared_loss

for epoch in range(num_epochs):
    for X, y in data_iter(batch, features, labels):
        with tf.GradientTape() as t:
            t.watch([w,b])
            l = loss(net(X, w, b), y)
        grads = t.gradient(l, [w, b])
        sgd([w, b], lr, batch, grads)
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, tf.reduce_mean(train_l)))
```

# 调用实现

```python
import tensorflow as tf

num_inputs = 2
num_examples = 1000
true_w = [2, -3.4]  # 实际参数
true_b = 4.2         # 实际
features = tf.random.normal(shape=(num_examples, num_inputs), stddev=1)
labels = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_b
labels += tf.random.normal(labels.shape, stddev=0.01)
```

```python
# 读取数据
from tensorflow import data as tfdata

batch_size = 10
# 将训练数据的特征和标签组合
dataset = tfdata.Dataset.from_tensor_slices((features, labels))
# 随机读取小批量
dataset = dataset.shuffle(buffer_size=num_examples) 
dataset = dataset.batch(batch_size)
data_iter = iter(dataset)
```

```python
for X, y in data_iter:
    print(X, y)
    break
```

```python
# 推荐写法
for (batch, (X, y)) in enumerate(dataset):
    print(X, y)
    break
```

```python
#定义模型和初始化参数

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow import initializers as init
model = keras.Sequential()
model.add(layers.Dense(1, kernel_initializer=init.RandomNormal(stddev=0.01)))

# 损失函数
from tensorflow import losses
loss = losses.MeanSquaredError()

# 优化算法
from tensorflow.keras import optimizers
trainer = optimizers.SGD(learning_rate=0.03)
```

```python
# 训练模型
num_epochs = 6
for epoch in range(1, num_epochs + 1):
    for (batch, (X, y)) in enumerate(dataset):
        with tf.GradientTape() as tape:
            l = loss(model(X, training=True), y)

        grads = tape.gradient(l, model.trainable_variables)
        trainer.apply_gradients(zip(grads, model.trainable_variables))

    l = loss(model(features), labels)
    print('epoch %d, loss: %f' % (epoch, l))
```
