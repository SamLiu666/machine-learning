```python
import re,jieba
```

```python
f = open('corpus.txt', 'r', encoding="utf-8")
text = f.read()
f.close()
```

```python
print(type(text))
text
```

# 分词

```python
segment = jieba.cut(text, cut_all=False, HMM=True)
```

```python
seg = " ".join(segment)
seg
```

```python
len(seg), type(seg)
```

```python
ff = open('segment.txt', 'w', encoding="utf-8")
ff.write(seg)
ff.close()
```

jieba 分词（HMM）， 存储到segment.txt 文件
