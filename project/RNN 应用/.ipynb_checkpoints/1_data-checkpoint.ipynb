```python
import request
import re
```

```python
url = 'https://www.shigeku.org/xlib/lingshidao/yishi/tagore.htm'
response = requests.get(url)
```

```python
content = response.text
print(content)
```

```python
type(content)
```

```python
pattern = re.compile(u'[\u4e00-\u9fa5]+')
text = " ".join(pattern.findall(content))
print(text)
type(text)
```

```python
len(text)
```

```python
f = open('corpus.txt', 'w', encoding='utf-8')
f.write(text)
f.close()
```
