```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(
    'https://labfile.oss.aliyuncs.com/courses/1283/telecom_churn.csv')  # 电信运营商实验数据
```

```python
df.head()
```

# 单变量可视化

## 直方图

```python
features = ["Total day charge", "Area code","Customer service calls"]
df[features].hist(figsize=(11,5))
print("\n")
features = ['Total day minutes', 'Total intl calls']
df[features].hist(figsize=(10,4))
```

## 密度图


```python
df[features].plot(kind='density', subplots=True, layout=(1,2), 
                  sharex=False,figsize=(10,4), legend=False, title=features)
```

同时展示直方图和密度图


```python
sns.distplot(df["Customer service calls"])
sns.distplot(df["Total intl calls"])
```
