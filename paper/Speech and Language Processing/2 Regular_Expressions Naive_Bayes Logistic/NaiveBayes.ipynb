```python
def read_data(file_name):
    data = []
    with open(file_name) as f:
        for line in f:
            (label, sentences) = line.split("\t")
            sentence_list = sentences.split('，')
            data.extend([(sentence, label) for sentence in sentence_list if sentence])
    return data
```

```python
data = read_data('1.txt')
```

```python
import jieba
def get_features(sentence):
    data = {}
#     seg_list = pesg.cut(sentence)
    for words, tag in sentence:
        word = jieba.lcut(words)
        if tag in data:
            data[tag] += word
        else:
            data[tag] = word
    return data
```

```python
def get_sets(datafile):
    features_sets = []
    data = read_data(datafile)
    for sentence, label in data:
        feature = get_features(sentence)
        features_sets.append((feature, label))
    return features_sets
```

朴素贝叶斯训练步骤：

2. 训练模型--生成式判别模型，直接输入测试数据，根据最大似然估计，概率大的为对应类别如果是未出现的词，+1 进行平滑处理
 - Blue，Red 概率比较，取值大的为判别的概率。
3. 识别结果100%

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
data_blue=[]
data_red=[]
word_count_blue = {}
word_count_red = {}

for line in data:
    word_list = line.split(" ")
    
    if word_list[0] == "BLUE":
        for i in range(1,len(word_list)):
            if word_list[i] not in stopwords:  # stopwords
                data_blue.append(word_list[i])
                
                # 统计单词出现次数
                word_count_blue.setdefault(word_list[i], 0)
                word_count_blue[word_list[i]] += 1
                
    if word_list[0] == "RED":
        for i in range(1,len(word_list)):
            if word_list[i] not in stopwords:  # stopwords
                data_red.append(word_list[i])
                
                # 统计单词出现次数
                word_count_red.setdefault(word_list[i], 0)
                word_count_red[word_list[i]] += 1
                
#统计非重复词
unique_words_blue = list(set(data_blue))
unique_words_red = list(set(data_red))
```

```python
test_data = []

for line in test:
    content_data = line.split(" ")
    #print(content_data)
    if content_data[0] == "BLUE":
        content_data.pop(0)
        test_data.append({"class": "blue", "content": content_data})
        
    #print(content_data)
    if content_data[0] == "RED":
        content_data.pop(0)
        test_data.append({"class": "red", "content": content_data})
```

```python
blue_ = []
red_ = []
for line_data in test_data:
    if line_data["class"] == "blue":
        posibility_blue = 0
        for word in line_data["content"]:
            # 计算每个单词的概率，取对数
            if word in word_count_blue:
                posibility_blue += math.log((word_count_blue[word]+1)/(len(data_blue)+len(unique_words_blue)+len(unique_words_red)))
            else:
                posibility_blue += math.log(1/(len(data_blue)+len(unique_words_blue)+len(unique_words_red)))
        blue_.append(posibility_blue)
        
    if line_data["class"] == "red":
        posibility_red = 0
        for word in line_data["content"]:
            # 计算每个单词的概率，取对数
            if word in word_count_red:
                posibility_red += math.log((word_count_red[word]+1)/(len(data_red)+len(unique_words_blue)+len(unique_words_red)))
            else:
                posibility_red += math.log(1/(len(data_red)+len(unique_words_blue)+len(unique_words_red)))
                
        red_.append(posibility_red)
```
