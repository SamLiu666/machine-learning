# Perceptron: The Simplest Neural Network
感知机 ![perception](https://nlp-pt.apachecn.org/docs/img/58bd03f5bc961bd29d7f05203ba2e97f.jpg)
每个感知器单元有一个输入(x),一个输出(y),和三个“旋钮”（knobs）:一组权重(w),偏量(b),和一个激活函数(f)

```python
import torch
import torch.nn as nn

class Perception(nn.Module):
    """感知机是一个线性层"""
    
    def __init__(self, input_dim):
        """inputr_dim: 输入样本的特征值"""
        
        super(Perception, self).__init__()
        self.fc1 = nn.Linear(input_dim, 1)
        
    def forward(self, x_in):
        """感知机前向传递：
        x_in: torch.Tensor 输入数据的张量， .shape =（batch, num_features）
        返回 结果张量，tensor.shape = (batch,)"""
        return torch.sigmoid(self.fc1(x_in)).squeeze()
```

# Activation Functions
激活函数 ![a](https://nlp-pt.apachecn.org/docs/img/3/%E6%BF%80%E6%B4%BB%E5%87%BD%E6%95%B0.png)
激活函数是神经网络中引入的非线性函数，用于捕获数据中的复杂关

## Sigmoid 
作用：将任何实值并将其压缩在0和1之间
$$f(x) = \frac{1}{{1 + {e^{ - x}}}}$$

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

x = torch.range(-8., 8., 0.1)
y = torch.sigmoid(x)

plt.plot(x.numpy(), y.numpy())
plt.title("Sigmoid Function")
plt.show()
```

## Tanh
$$f(x) = \frac{e^x - e^{(-x)}}{{e^x + e^{(-x)}}}$$


```python
y = torch.tanh(x)

plt.plot(x.numpy(), y.numpy())
plt.title("Tanh Function")
plt.show()
```

 可以看到，sigmoid 和 tanh 比较相似，具体区别如下:
1、Sigmoid函数比Tanh函数收敛饱和速度慢

2、Sigmoid函数比Tanh函数梯度收敛更为平滑

3、Sigmoid函数比Tanh函数值域范围更窄

函数值变化敏感，一旦接近或者超出区间就失去敏感性，处于饱和状态，影响神经网络预测的精度值。tanh的输出和输入能够保持非线性单调上升和下降关系，符合BP网络的梯度求解，容错性好，有界，渐进于0、1，符合人脑神经饱和的规律，但比sigmoid函数延迟了饱和期。

# Relu
$$f(x)=max(0,x)$$

```python
y = torch.relu(x)
plt.plot(x.numpy(), y.numpy())
plt.title("Relu Function")
plt.show()
```

ReLU的裁剪效果有助于消除梯度问题，随着时间的推移，网络中的某些输出可能会变成零，再也不会恢复。这就是所谓的“dying ReLU”问题。为了减轻这种影响，提出了Leaky ReLU或 Parametric ReLU (PReLU)等变体，其中泄漏系数a是一个可学习参数: 
    $$f(x)=max(x,ax)$$

```python
prelu = torch.nn.PReLU(num_parameters=1)
y = prelu(x)

plt.plot(x.numpy(), y.detach().numpy())
plt.title("PRelu Function")
plt.show()
```

# Softmax


```python
softmax = nn.Softmax()
y = softmax(x)

plt.plot(x.numpy(), y.numpy())
plt.title("Relu Function")
plt.show()
```

```python
softmax = nn.Softmax(dim=1)
x_input = torch.randn(1, 3)
y_output = softmax(x_input)
print(x_input)
print(y_output)
print(torch.sum(y_output, dim=1))

plt.scatter(x_input.numpy(), y_output.numpy())
plt.title("Relu Function")
plt.show()
```

# Loss Function 
损失函数

## Mean Squared Error Loss 
MSE： 均方误差,预测值与目标值之差的平方的平均值
平均绝对误差(MAE)和均方根误差(RMSE)，但是它们都涉及到计算输出和目标之间的实值距离

```python
mse_loss = nn.MSELoss()
outputs = torch.randn(3, 5, requires_grad=True)
targets = torch.randn(3, 5)
loss = mse_loss(outputs, targets)
print(loss)
```

## Categorical Cross-Entropy Loss
分类交叉熵损失(categorical cross-entropy loss)通常用于多类分类设置，其中输出被解释为类隶属度概率的预测。


```python
ce_loss = nn.CrossEntropyLoss()
outputs = torch.randn(3, 5, requires_grad=True)
targets = torch.tensor([1, 0, 3], dtype=torch.int64)
loss = ce_loss(outputs, targets)
print(loss)
```

## Binary Cross-Entropy
二元交叉熵损失: 表示网络输出的随机向量上的sigmoid激活函数创建二进制概率输出向量。接下来，ground truth被实例化为一个0和1的向量。最后，利用二元概率向量和基真值向量计算二元交叉熵损失。

```python
bce_loss = nn.BCELoss()
sigmoid = nn.Sigmoid()
probabilities = sigmoid(torch.randn(4, 1, requires_grad=True))

targets = torch.tensor([1, 0, 1, 0],  dtype=torch.float32).view(4, 1)
loss = bce_loss(probabilities, targets)

print(probabilities)
print(loss)
```

# Diving Deep into Supervised Training
监督学习：学习如何将观察结果映射到特定目标的问题，如何使用模型预测和损失函数对模型参数进行基于梯度的优化

模型、损失函数、训练数据和优化算法。监督学习的训练数据是观察和目标对，模型从观察中计算预测，损失衡量预测相对于目标的误差。训练的目的是利用基于梯度的优化算法来调整模型的参数，使损失尽可能小。

举例：玩具问题:将二维点划分为两个类中的一个。直观上，这意味着学习一条直线(称为决策边界或超平面)来区分类之间的点。我们一步一步地描述数据结构，选择模型，选择一个损失，建立优化算法，最后，一起运行它。
